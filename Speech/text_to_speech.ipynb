<h1> Importing Libraries

```python
import os
from gtts import gTTS 
import librosa
import soundfile as sf
import IPython
```

<h2> Function to Convert Text to Speech <h2>

```python
def text_to_speech(st,file_name):
  language='en'
  obj=gTTS(st,lang=language,slow=False,lang_check=True)
  print(file_name)
  obj.save(file_name)
  x,_=librosa.load(file_name,sr=16000)
  sf.write(file_name,x,16000)
```

```python
st="I love to fuck bitches"
text_to_speech(st,'/content/st1.wav')
```

```python
IPython.display.Audio('/content/st1.wav')
```

```python
st2="I have to say that Biswajit Jagadev is so rich that everyone calls him daddy"
text_to_speech(st2,'/content/jaga.wav')
```
