```python
df=pd.read_excel('../input/project-data-audio-files-and-csv/database info.xlsx',sheet_name='Sayfa1')
df.head()
```

```python
import librosa
import IPython.display as id
import scipy.io.wavfile as wave
```

```python
files=df['FILE NAME'].values
mfcc=[]
sample_rate=[]
for i in files:
    path='../input/project-data-audio-files-and-csv/'+i
    #print(path)
    audio,rate=librosa.load(path,res_type='kaiser_fast')
    sam_rate,un_normalised_audio=wave.read(path)
    sample_rate.append(sam_rate)
    mfcc_val=librosa.feature.mfcc(y=audio,sr=rate,n_mfcc=40)
    normalised_mfcc=np.mean(mfcc_val)
    mfcc.append(normalised_mfcc)
```

```python
df['Sample_rate']=sample_rate
df['mfcc_coeficient']=mfcc
df.head()
```

## Regression using Pytorch

```python
import torch
from torch.autograd import Variable
X=df['mfcc_coeficient'].values
y1=df['SYSTOLİC BP'].values
y2=df['DİASTOLİC BP'].values
print(X.shape,y1.shape,y2.shape)
```

```python
class LinearModel(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.linear=torch.nn.Linear(1,1)
    def forward(self,x):
        y_pred=self.linear(x)
        return y_pred
#print(X,y1,y2)
X_train=[[i] for i in X]
y1_train=[[i] for i in y1]
y2_train=[[i] for i in y2]
X=Variable(torch.Tensor(X_train))
y1=Variable(torch.Tensor(y1_train))
y2=Variable(torch.Tensor(y2_train))
```

```python
model=LinearModel()
criterion = torch.nn.MSELoss()
print(model.parameters())
optimizer = torch.optim.SGD(model.parameters(), lr = 0.01)
for epoch in range(2000):
    y1_pred=model.forward(X)
    loss=criterion(y1_pred,y1)
    if(epoch%5==0):
        print('epoch {}, loss {}'.format(epoch, loss.item()))
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
model2=LinearModel()
criterion = torch.nn.MSELoss()
print(model2.parameters())
optimizer = torch.optim.SGD(model2.parameters(), lr = 0.01)
for epoch in range(1500):
    y2_pred=model2.forward(X)
    loss=criterion(y2_pred,y2)
    if(epoch%5==0):
        print('epoch {}, loss {}'.format(epoch, loss.item()))
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```
