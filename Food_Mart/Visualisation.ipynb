# Visualisation
## Vivid Data Visualisation and EDA of the Training Set

```
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
```

```
df_train=pd.read_csv("/content/training_set.csv")
df_train.head()
```

```
sns.set_theme(style='darkgrid')
sns.countplot(x='cuisine',data=df_train,color='cyan')
```

```
sns.set_theme(style='darkgrid')
sns.countplot(x='category',data=df_train,color='cyan')
```

```
plt.plot(df_train.op_area)
plt.show()
```

```
sns.set_theme(style='darkgrid')
sns.countplot(x='center_type',data=df_train,color='cyan')
```

```
sns.countplot(x='center_type',hue='cuisine',data=df_train,color='cyan')
```

```
sns.countplot(x='center_type',hue='category',data=df_train,color='cyan')
```

```
sns.countplot(x='center_type',hue='emailer_for_promotion',data=df_train,color='cyan')
```

```
sns.countplot(x='emailer_for_promotion',data=df_train,color='cyan')
```

```
sns.countplot(x='center_type',hue='homepage_featured',data=df_train,color='cyan')
```

```
sns.countplot(x='homepage_featured',data=df_train,color='cyan')
```

```
dfA=df_train[df_train.center_type=='TYPE_A']
plt.scatter(x=dfA['Unnamed: 0'],y=dfA['base_price'],color='cyan')
plt.ylabel('Base Price')
plt.title('Price Distribution for Center Type A')
plt.show()
```

```
dfB=df_train[df_train.center_type=='TYPE_B']
plt.scatter(x=dfB['Unnamed: 0'],y=dfB['base_price'],color='cyan')
plt.ylabel('Base Price')
plt.title('Price Distribution for Center Type B')
plt.show()
```

```
dfC=df_train[df_train.center_type=='TYPE_C']
plt.scatter(x=dfC['Unnamed: 0'],y=dfC['base_price'],color='cyan')
plt.ylabel('Base Price')
plt.title('Price Distribution for Center Type C')
plt.show()
```

```
g=sns.PairGrid(df_train, vars=["base_price", "checkout_price"], hue="center_type")
g.map(sns.scatterplot)
```
