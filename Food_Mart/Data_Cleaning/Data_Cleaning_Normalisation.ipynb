```python
import pandas as pd 
import numpy as np
import sklearn
```

```python
df_train=pd.read_csv('/content/training_set.csv')
df_train.head()
```

```python
df_test=pd.read_csv('/content/testing_set.csv')
df_test.head()
```

```python
df_train.drop(['Unnamed: 0'],axis=1,inplace=True)
df_test.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
lb1=LabelEncoder()
k1=df_train['category']
k2=df_test['category']
lb1.fit(k1)
k1=lb1.transform(k1)
k2=lb1.transform(k2)
df_train['category']=k1
df_test['category']=k2
```

```python
lb2=LabelEncoder()
k1=df_train['cuisine']
k2=df_test['cuisine']
lb2.fit(k1)
k1=lb2.transform(k1)
k2=lb2.transform(k2)
df_train['cuisine']=k1
df_test['cuisine']=k2
```
